# Creating/Updating the bronze tables

```python
for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} → workspace.bronze.{item['table']}")

    df = (
        spark.read
             .option("header", "true")
             .option("inferSchema", "true")
             .csv(item["path"])
    )

    (
        df.write
          .mode("overwrite")
          .format("delta")
          .saveAsTable(f"workspace.bronze.{item['table']}")
    )
```

## Sanity checks for bronze tables

```python
for item in INGESTION_CONFIG:
    print(f"Checking {item['source']} → workspace.bronze.{item['table']}")

    df = spark.read.table(f"workspace.bronze.{item['table']}")

    assert df.count() > 0, f"Table {item['table']} is empty"
    
```
